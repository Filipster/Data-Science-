#### Separando os dados em duas parte, profissionais e pacientes

```python
professionals = df[['Unidad_Operativa', 'Lugar_atención', 'Apellidos_profesional', 'Especialidad_subespecialidad_profesional', 'Sexo_profesional', 
                    'Fecha_nacimiento_profesional', 'Nacionalidad_profesional']].copy()

patients = df[['Unidad_Operativa', 'Lugar_atención', 'Apellidos_paciente', 
               'Sexo_paciente', 'Fecha_nacimiento_paciente', 'Nacionalidad_paciente', 
               'Autoidentificación_étnica_paciente', 'Afiliación_paciente', 'Grupo_prioritario_1', 
               'Grupo_prioritario_2', 'Grupo_prioritario_3', 'Residencia_provincia_paciente', 
               'Residencia_cantón_paciente', 'Residencia_parroquia_paciente', 'Condición_del_diagnóstico_a', 
               'Procedimiento_1', 'Orientación_Sexual', 'sheet_name', 
               'date_atencion', 'CID', 'DISEASE']].copy()
```

#### Feature engineering

```python
professionals['Fecha_nacimiento_profesional'] = pd.to_datetime(professionals['Fecha_nacimiento_profesional'])
patients['Fecha_nacimiento_paciente'] = pd.to_datetime(patients['Fecha_nacimiento_paciente'])
patients['date_atencion'] = pd.to_datetime(patients['date_atencion'])
```

```python
professionals['edad'] = professionals['Fecha_nacimiento_profesional'].apply(lambda time: int(datetime.now().strftime('%Y')) - time.year)
patients['edad'] = patients['Fecha_nacimiento_paciente'].apply(lambda time: int(datetime.now().strftime('%Y')) - time.year)
```

```python
patients['mes'] = patients['date_atencion'].apply(lambda time: time.month)
patients['dia_semana'] = patients['date_atencion'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0: 'lunes', 1: 'martes', 2: 'miercoles', 3: 'jueves', 4: 'viernes', 5: 'sabado', 6: 'domingo'}
mmap = {1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto', 9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre'}

patients['dia_semana'] = patients['dia_semana'].map(dmap)
patients['mes'] = patients['mes'].map(mmap)
```

**Los datos estadísticos que estan en el documento deben ser analizados deacuerdo a los grupos etarios definidos ejm. Menores de 1 año de 1-4años de 5-9, 10-14,15-19,20-64 y mayores de 65 años**

```python
def age_range(n):
    if n < 1:
        'menores de 1 año'
    elif 1 <= n <= 4:
        return '1-4 años'
    elif 5 <= n <= 9:
        return '5-9 años'
    elif 10 <= n <= 14:
        return '10-14 años'
    elif 15 <= n <= 19:
        return '15-19 años'
    elif 20 <= n <= 64:
        return '20-64 años'
    else:
        return 'mayores de 65 años'

professionals['edad_extensión'] = professionals['edad'].apply(age_range)
patients['edad_extensión'] = patients['edad'].apply(age_range)
```

```python
def annt(rects, value):
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{:.2f}%'.format(height / value * 100),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```
